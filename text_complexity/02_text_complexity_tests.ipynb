```python
import pandas as pd
import scipy as sc
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu
```

```python
df2=pd.read_csv('df94.csv')
```

```python
df2.periodical_title.unique()
```

## Bastide

```python
df=df2.loc[df2.author=='Jean-François de Bastide']

df.boxplot(by='periodical_title',column='flesch_reading_ease')

x=df.loc[df.periodical_title=='Le Monde'].flesch_reading_ease.values
y=df.loc[df.periodical_title=='Le Monde comme il est'].flesch_reading_ease.values
z=df.loc[df.periodical_title=='Le Nouveau Spectateur'].flesch_reading_ease.values
print(mannwhitneyu(x, y,alternative='less'))
print(mannwhitneyu(x, z,alternative='less'))
print(mannwhitneyu(y, z,alternative='less'))
```

```python
df.boxplot(by='periodical_title',column='delta_TTR')

x=df.loc[df.periodical_title=='Le Monde'].delta_TTR.values
y=df.loc[df.periodical_title=='Le Monde comme il est'].delta_TTR.values
z=df.loc[df.periodical_title=='Le Nouveau Spectateur'].delta_TTR.values
print(mannwhitneyu(x, y,alternative='less'))
print(mannwhitneyu(x, z,alternative='less'))
print(mannwhitneyu(y, z,alternative='less'))
```

```python
df.boxplot(by='periodical_title',column='p3/p1')

x=df.loc[df.periodical_title=='Le Monde']['p3/p1'].values
y=df.loc[df.periodical_title=='Le Monde comme il est']['p3/p1'].values
z=df.loc[df.periodical_title=='Le Nouveau Spectateur']['p3/p1'].values
print(mannwhitneyu(x, y,alternative='greater'))
print(mannwhitneyu(x, z,alternative='greater'))
print(mannwhitneyu(y, z,alternative='greater'))
```

```python
import numpy as np
def IQR(x):
    return np.percentile(x.values,75)-np.percentile(x.values,25)
```

```python
df=df2.groupby('periodical_title').agg({'author':'first','language':'first','filename':'count','word_len':'mean',
                                    'flesch_reading_ease':['mean','std'],'delta_TTR':['mean','std'],
                                    'p3/p1':['mean','std']})
df=df.sort_values(by=('author', 'first'))
df=df.drop('Le Spectateur français avant la Révolution')
```

```python
df['Flesch'] = df[('flesch_reading_ease','mean')].round(1).apply(str)
df['Flesch'] = df['Flesch'] + ' (' + df[('flesch_reading_ease','std')].round(1).apply(str) + ')'
df=df.drop('flesch_reading_ease',axis=1)
```

```python
df['Delta TTR'] = df[('delta_TTR','mean')].round(3).apply(str)
df['Delta TTR'] = df['Delta TTR'] + ' (' + df[('delta_TTR','std')].round(3).apply(str) + ')'
df=df.drop('delta_TTR',axis=1)
```

```python
df['PoS-Rate'] =  df[('p3/p1','mean')].round(3).apply(str)
df['PoS-Rate'] = df['PoS-Rate'] + ' (' + df[('p3/p1','std')].round(3).apply(str) + ')'
df=df.drop('p3/p1',axis=1)
```

```python
print(df.to_latex(float_format="%.3f"))
```
