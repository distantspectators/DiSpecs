# Evaluation of the manually assigned keywords in the Spectators (part 1)

## Get metadata from XML files

Extract the periodical titles, ids, languages and manually assigned keywords from the XML files.   
 

```python
# create an empty pandas data frame
df = pd.DataFrame(columns=["id","periodical","lang","keywords"])
df
```

```python
# TEST code on one file before using on the whole corpus
with open('Y:/data/projekte/dispecs/topicModeling/2021-03-16/mws-097-349.xml', encoding="utf-8") as f:
    soup = BeautifulSoup(f, 'xml')
    for pid in soup.find('idno', type="PID"):
        pid = pid
    for periodical in soup.find('title', level='j'): 
        periodical = " ".join(periodical.split())
    for lang in soup.find('language'):
        lang = lang 
    term_group = []
    for term_set in soup.find_all('term', attrs={"xml:lang":"en"}):
        term_group.append(" ".join(term_set.string.split())) 
    
    values_to_add = {'id': pid, 'periodical': periodical, 'lang':lang, 'keywords':"; ".join(term_group)}
    row_to_add = pd.Series(values_to_add)
    df = df.append(row_to_add, ignore_index=True)

df
```

```python
# clear the data frame 
df = df.iloc[0:0]
```

```python
# fill the dataframe with the values from the xml files
for path in path_to_folder.glob('*.xml'):
    with open(path, encoding="utf-8") as f:
        soup = BeautifulSoup(f, 'xml')    
        if soup.find('term') != None: 
            term_group = []
            for term_set in soup.find_all('term', attrs={"xml:lang":"en"}):
                term_group.append(" ".join(term_set.string.split()))  
            for periodical in soup.find('title', level='j'): 
                periodical = " ".join(periodical.split())   
            for pid in soup.find('idno', type="PID"):
                pid = pid
            for lang in soup.find('language'):
                lang = lang
            values_to_add = {'id': pid, 'periodical': periodical, 'lang':lang, 'keywords':"; ".join(term_group)}
            row_to_add = pd.Series(values_to_add)
            df = df.append(row_to_add, ignore_index=True)

df            
```

```python
# convert the value types to string in case they have type "dataframe bs4.element.navigablestring"
df["id"] = df["id"].astype(str) 
df["lang"] = df["lang"].astype(str)
```

```python
# save the data frame to a pickle file
df.to_pickle(output+"file_infos.pkl")
```
